### Missing Values

ตรวจสอบว่าข้อมูลที่อยู่ในไฟล์ csv มี missing value หรือไม่

```python
air.isnull()
```

```python
air.columns
```

นับจำนวนค่าที่ไม่เป็น missing value

```python
air.dropna().sum()
```

เลือกวิธีการที่เหมาะสมสำหรับจัดการกับค่า missing value

```python
air_avg = air["minimum nights"].mean()
night_nomiss = air['minimum nights'].fillna(air_avg)
night_nomiss
```

```python
air_avg = air["reviews per month"].mean()
month_nomiss = air["reviews per month"].fillna(air_avg)
month_nomiss
```

```python
air_avg = air["review rate number"].mean()
num_nomiss = air["review rate number"].fillna(air_avg)
num_nomiss
```

```python
air_avg = air["availability 365"].mean()
avi_nomiss = air["availability 365"].fillna(air_avg)
avi_nomiss
```
