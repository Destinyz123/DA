```python
import pandas as pd
import statsmodels.formula.api as sm
from sklearn import linear_model

import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
# อ่านข้อมูลจากไฟล์ excel
insur = pd.read_csv('insurance.csv')
insur.head()
```

**Columns**

**age**: age of primary beneficiary

**sex**: insurance contractor gender, female, male

**bmi**: Body mass index, providing an understanding of body, weights that are relatively high or low relative to height,
objective index of body weight (kg / m ^ 2) using the ratio of height to weight, ideally 18.5 to 24.9

**children**: Number of children covered by health insurance / Number of dependents

**smoker**: Smoking

**region**: the beneficiary's residential area in the US, northeast, southeast, southwest, northwest.

**charges**: Individual medical costs billed by health insurance


## Questions
1. ความสัมพันธ์ระหว่างค่ารักษาพยาบาลและคุณสมบัติต่างๆสามารถวิเคราะห์ได้อย่างไร? สามารถคาดการณ์ผลกระทบของแต่ละคุณสมบัติที่มีผลต่อมูลค่าค่ารักษาพยาบาลได้หรือไม่? ความสัมพันธ์เป็นเส้นตรงหรือไม่
+  คุณสมบัติใด (age, bmi, children,) ที่ทำให้ค่าใช้จ่ายมากที่สุด
+  ตัวแปรแต่ละตัวมีความสัมพันธ์กันหรือไม่


```python
#Ordinary Least Squares: OLS เป็นการประมาณค่าด้วยวิธีกําลังสองน้อยที่สุดทําให้ผลรวมกําลังสองของค่า error น้อยที่สุด
# 'age', 'charges'
ad_Age_ols = sm.ols(formula="charges ~ age", data=insur) 
Age_fit_model = ad_Age_ols.fit() # ปรับ model regression เพื่อให้ค่า error น้อยที่สุด
Age_fit_model.summary()
```

```python
# ใช้ scatter plot เพื่อดูความสัมพันธ์ของตัวแปร และแสดงเส้นพยากรณ์ของความสัมพันธ์ของ 2 ตัวแปร
# 'age', 'charges'
lr = linear_model.LinearRegression() # create a linear regression object

# scikit-learn doesn't work as well with pandas, so we have to extract values 
x = insur['age'].values.reshape(insur['age'].shape[0],1)
y = insur['charges'].values.reshape(insur['charges'].shape[0],1)

lr.fit(X=x, y=y)

plt.scatter(x, y,  color='black')
plt.plot(x, lr.predict(x), color='blue', linewidth=3)
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

## ต้องการวิเคราะห์ว่าตัวแปรอื่นๆ มีผลต่อมูลค่ายอดขายสินค้าหรือไม่
ใช้ linear regression วิเคราะห์ค่าใช้จ่ายในการโฆษณาทางวิทยุและหนังสือพิมพ์มีผลต่อมูลค่ายอดขายสินค้าหรือไม่ โดยใช้ statsmodels

```python
#  'bmi', 'charges'
# ปรับ model regression เพื่อให้ค่า error น้อยที่สุด
ad_bmi_ols = sm.ols(formula="charges ~ bmi", data=insur) 
bmi_fit_model = ad_bmi_ols.fit() # ปรับ model regression เพื่อให้ค่า error น้อยที่สุด
bmi_fit_model.summary()
```

```python
# ใช้ scatter plot เพื่อดูความสัมพันธ์ของตัวแปร และแสดงเส้นพยากรณ์ของความสัมพันธ์ของ 2 ตัวแปร
# 'bmi', 'charges'
lr = linear_model.LinearRegression() # create a linear regression object

# scikit-learn doesn't work as well with pandas, so we have to extract values 
x = insur['bmi'].values.reshape(insur['bmi'].shape[0],1)
y = insur['charges'].values.reshape(insur['charges'].shape[0],1)

lr.fit(X=x, y=y)

plt.scatter(x, y,  color='black')
plt.plot(x, lr.predict(x), color='blue', linewidth=3)
plt.xlabel('Bmi')
plt.ylabel('Charges')
plt.show()
```

```python
#'children', 'charges'
# ปรับ model regression เพื่อให้ค่า error น้อยที่สุด
ad_children_ols = sm.ols(formula="charges ~ children", data=insur) 
children_fit_model = ad_children_ols.fit() # ปรับ model regression เพื่อให้ค่า error น้อยที่สุด
children_fit_model.summary()
```

```python
# ใช้ scatter plot เพื่อดูความสัมพันธ์ของตัวแปร และแสดงเส้นพยากรณ์ของความสัมพันธ์ของ 2 ตัวแปร
# 'children', 'charges'
lr = linear_model.LinearRegression() # create a linear regression object

# scikit-learn doesn't work as well with pandas, so we have to extract values 
x = insur['children'].values.reshape(insur['children'].shape[0],1)
y = insur['charges'].values.reshape(insur['charges'].shape[0],1)

lr.fit(X=x, y=y)

plt.scatter(x, y,  color='black')
plt.plot(x, lr.predict(x), color='blue', linewidth=3)

plt.xlabel('children')
plt.ylabel('Charges')
plt.show()
```

```python
#ใช้ scatter plot เพื่อดูความสัมพันธ์ของตัวแปร'age', 'bmi', 'children', 'charges'และแสดงเส้นพยากรณ์ของความสัมพันธ์ของ 4 ตัวแปร

plt.scatter(x=insur['age'],y=insur['charges'],c='r',marker='s',label='age')
plt.scatter(x=insur['bmi'],y=insur['charges'],c='b',marker='o',label='bmi')
plt.scatter(x=insur['children'],y=insur['charges'],c='g',marker='*',label='children')

#
#
plt.legend(numpoints=1,loc=4)

plt.plot(insur['age'],Age_fit_model.predict(),c='r',linewidth=3)
plt.plot(insur['bmi'],bmi_fit_model.predict(),c='b',linewidth=3)
plt.plot(insur['children'],children_fit_model.predict(),c='k',linewidth=3)
#
#

plt.xlabel('Ad budget (Thousands of dollars)')
plt.ylabel('Sales (units of product)')
plt.show()
```
